```python
import numpy as np
import matplotlib.pyplot as plt
from array import array       # PYTHON STANDARD LIBRARY ARRAY
%matplotlib inline
```

```python
elev = array('B')             # B FOR INT SIZE 1 BYTE AS 1-byte elevation

with open("colorado_elev.vit", "rb") as data:
    header = data.read(268)
    elev.fromfile(data, 160000)
    file_size = data.tell()
print("file size:" ,file_size)
#print("Header:\n",header)
#print("Elevation Data in Array:\n" ,elev)
```

```python
array = np.array(elev)
print(array)
print(array.shape)
```

```python
array_2d = array.reshape(400,400)
print(array_2d)
print(array_2d.shape)
```

```python
plt.imshow(array_2d, cmap='terrain')
plt.colorbar()
plt.title("Image: Terrain")
plt.savefig("elevation_terrain.jpg")
```

```python
plt.imshow(array_2d, cmap='hsv')
plt.colorbar()
plt.title("Image: HSV")
plt.savefig("elevation_hsv.jpg")
```

```python
plt.imshow(array_2d, cmap='gray')
plt.colorbar()
plt.title("Image: Grayscale")
plt.savefig("elevation_grayscale.jpg")
```

```python
plt.plot(array_2d[100], label = "line 100")
plt.plot(array_2d[200], label = "line 200")
plt.plot(array_2d[300], label = "line 300")
plt.xlabel("pixels")
plt.ylabel("Elevation")
plt.title("Elevation of different profile lines")
plt.legend()
plt.savefig("elevation_lines.jpg")
plt.show()
```

```python
print("Minimum Value:",np.min(array_2d))
print("Maximum Value:",np.max(array_2d))
```
